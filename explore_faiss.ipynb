```python
import numpy as np
```

```python
file_directory = "10k_text/"
ticker = 'ABCL'

# read the .txt file
file_path = file_directory + ticker + ".txt"

# read the file
with open(file_path, 'r', encoding='utf-8') as file:
    text = file.read()

print(f"File {file_path} has {len(text)} characters")
```

```python
from transformers import AutoTokenizer

CHUNK_SIZE = 1024
OVERLAP = 256

tokenizer = AutoTokenizer.from_pretrained("sujet-ai/Marsilia-Embeddings-EN-Base", use_fast=True)
```

```python
# Tokenize full text and split into overlapping chunks
token_ids = tokenizer.encode(text, add_special_tokens=False)

stride = CHUNK_SIZE - OVERLAP
chunks = []
for start in range(0, len(token_ids), stride):
    chunk_ids = token_ids[start : start + CHUNK_SIZE]
    chunks.append(tokenizer.decode(chunk_ids, skip_special_tokens=True))
    if start + CHUNK_SIZE >= len(token_ids):
        break

print(f"Total tokens : {len(token_ids)}")
print(f"Total chunks : {len(chunks)}")
```

```python
from sentence_transformers import SentenceTransformer
model = SentenceTransformer("sujet-ai/Marsilia-Embeddings-EN-Base")
```

```python
embeddings = model.encode(chunks, show_progress_bar=True, batch_size=32)

print(f"{len(chunks)} chunks -> embeddings shape: {embeddings.shape}")
```

```python
query = "What are the drugs they are working on?"
query_embed = model.encode(query)

```

```python
from sklearn.metrics.pairwise import cosine_similarity

scores = cosine_similarity(query_embed.reshape(1, -1), embeddings).flatten()
top_5 = np.argsort(scores)[::-1][:10]

for rank, idx in enumerate(top_5, 1):
    print(f"--- Rank {rank} | Chunk {idx} | Score: {scores[idx]:.4f} ---")
    print(chunks[idx], "\n")
```
